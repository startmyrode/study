```python
#绘制散点图
# 导入必要的库
import pandas as pd
import matplotlib.pyplot as plt

# 创建数据
data = {'diameter': [6, 8, 10, 14, 18], 'value': [7, 9, 13, 17.5, 18]}
data_frame = pd.DataFrame(data)

# 绘制散点图
plt.scatter(data_frame['diameter'], data_frame['value'])

# 添加标题和标签
plt.title('diameter vs value')
plt.xlabel('diameter')
plt.ylabel('value')

# 显示图形
plt.show()
```

```python
#绘制线性回归模型，给出拟合直线方程
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 创建示例数据
data = {'diameter': [6, 8, 10, 14, 18], 'value': [7, 9, 13, 17.5, 18]}
data_frame = pd.DataFrame(data)

# 提取特征和目标变量
X = data_frame[['diameter']]  # 特征
y = data_frame['value']    # 目标变量

# 创建线性回归模型并拟合数据
model = LinearRegression()
model.fit(X, y)

# 预测值
y_pred = model.predict(X)

# 绘制散点图
plt.scatter(X, y, color='blue', label='real')

# 绘制拟合直线
plt.plot(X, y_pred, color='red', label='dimmy')

# 添加标签和标题
plt.xlabel('diameter')
plt.ylabel('value')
plt.title('xianxinghuiguinihe')
plt.legend()

# 显示图形
plt.show()
```

```python
X_test = pd.DataFrame({'diameter': [12]})
predicted_price = model.predict(X_test)
print(predicted_price)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 设置 Matplotlib 使用支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 导入数据
data = {'直径': [6, 8, 10, 14, 18], '价格': [7, 9, 13, 17.5, 18]}
data_frame = pd.DataFrame(data)

# 绘制散点图
plt.scatter(data_frame['直径'], data_frame['价格'])
plt.xlabel('直径')
plt.ylabel('价格')
plt.title('披萨直径与价格的关系')
plt.show()

# 训练线性回归模型
X = data_frame[['直径']]
y = data_frame['价格']
model = LinearRegression()
model.fit(X, y)

# 绘制拟合直线
plt.scatter(data_frame['直径'], data_frame['价格'], label='数据点')
plt.plot(X, model.predict(X), color='red', label='拟合直线')
plt.xlabel('直径')
plt.ylabel('价格')
plt.title('披萨直径与价格的线性回归')
plt.legend()
plt.show()

# 预测 12 英寸披萨的价格
predicted_price = model.predict(pd.DataFrame({'直径': [12]}))
print(f'预测的 12 英寸披萨价格: {predicted_price[0]:.2f}')

# 评价模型准确率
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'均方误差 (MSE): {mse:.2f}')
print(f'决定系数 (R²): {r2:.2f}')
```
